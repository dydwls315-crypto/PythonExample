```python
def fibonacci(n):
 if n <= 1:
  return n
 else:
  return fibonacci(n - 1) + fibonacci(n - 2)
print(fibonacci(7))
```

```python
def find_max(numbers):
 if len(numbers) == 1:
  return numbers[0]
 else:
  max_of_rest = find_max(numbers[1:])
 return numbers[0] if numbers[0] > max_of_rest else max_of_rest
my_list = [3, 7, 2, 9, 1]
print(find_max(my_list))
```

```python
def simple_gen():
 yield "Emil"
 yield "Tobias"
 yield "Linus"
gen = simple_gen()
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
def fibonacci():
 a, b = 0, 1
 while True:
  yield a
 a, b = b, a + b
# Get first 100 Fibonacci numbers
gen = fibonacci()
for _ in range(100):
 print(next(gen))
```

```python
def echo_generator():
 while True:
  received = yield
 print("Received:", received)
gen = echo_generator()
next(gen) # Prime the generator
gen.send("Hello")
gen.send("World")
```
